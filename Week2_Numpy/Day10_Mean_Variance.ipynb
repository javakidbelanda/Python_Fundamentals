```python
def mean(data):
    return sum(data)/len(data)
```

```python
def variance(data):
    m = mean(data)
    return sum((x-m)**2 for x in data)/(len(data)-1)
```

```python
def std_dev(data):
    return variance(data)**0.5
```

```python
data = [4, 7, 10, 10, 13]
```

```python
print("Mean:", mean(data))
print("Variance:", variance(data))
print("Std Dev:", std_dev(data))
```

```python
import numpy as np
np.mean(data)
np.var(data, ddof=1)
np.std(data, ddof=1)
```
