```python
def is_even(number):
    if number % 2 == 0:
        return True
    else:
        return False
```

```python
print (is_even(4))
print (is_even(7))
```

```python
for even in range(1,21):
    if is_even (even):
        print(even, "is even number")
```

```python
def describe_list (numbers):
    n = len(numbers)
    mean = sum(numbers) / n
    sorted_numbers = sorted(numbers)
    median = sorted_numbers[n//2]
    variance = sum((x-mean)**2 for x in numbers) / n
    std = variance ** 0.5
    return mean, median, std
print (describe_list([20,30,40]))
```

```python
student = {"Alice" : 82, "Bob" : 90, "Charlie" : 78, "David" : 85, "Eve" : 88}
for name, score in student.items():
    if score > 85:
        print(name, "has a good score")  
```

```python
import numpy as np

# Task 4: Create array of 50 random numbers between 0 and 100
arr = np.random.randint(0, 100, 50)
print("Array:", arr)
print("Mean:", arr.mean())
print("Std", arr.std())
print("Max:", arr.max())
print("Min:", arr.min())

# Task 5 : Reshape into 5x10 matrix
matrix = arr.reshape(5, 10)
print("Shaped Matrix:", matrix.shape)
print("Shaped Array:", matrix)

# Task 6 : Select all values greater than 60
greater_than_60 = arr[arr > 60]
print("Values greater than 60:", greater_than_60)
```
